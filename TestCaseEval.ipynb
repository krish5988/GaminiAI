```python
from langchain.prompts import PromptTemplate,ChatPromptTemplate,HumanMessagePromptTemplate,MessagesPlaceholder
from langchain_core.messages import SystemMessage
from langchain_experimental.agents.agent_toolkits import create_python_agent
from langchain_experimental.tools.python.tool import PythonREPLTool
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import LLMChain
```

```python
prompt = ChatPromptTemplate.from_messages([
    
    SystemMessage(content="Output data in tabular format"),
    # MessagesPlaceholder(variable_name="chat_hist"),
    HumanMessagePromptTemplate.from_template("create test cases in tabular format with column Testcase ID,test case description,Test Steps,Input Data,Expected result with maximum details from - {content}")
    
    ]
)
llm1=ChatGoogleGenerativeAI(model="gemini-pro",temperature=0.9)
chain=LLMChain(llm=llm1,
               prompt=prompt,
               verbose=True)
```

```python

from langchain.document_loaders.text import TextLoader
loader = TextLoader("./TestScenario.txt",encoding='utf-8');

docs = loader.load();
docs[0].page_content
```

```python
resp = chain.run({'content':docs[0].page_content})
print(resp)
```
